## Заполнение пропусков

Пропуски в этом датасете обозначены как `?`

```python
for c in df.columns:
    print(c, len(df[df[c] == '?']))
```

Удалите строки, для которых неизвестно значение price, так как это целевая переменная.

```python
df_new = df[df['price'] != '?']
df_new.shape
```

Заполните средним значением пропуски в столбцах для числовых признаков и самым популярным значением для категориальных признаков
* `num-of-doors`
* `bore`
* `stroke`
* `horsepower`
* `peak-rpm`

```python
data_new = df_new.copy()
data_new
numeric_features = ['bore', 'stroke', 'horsepower', 'peak-rpm']
for feature in numeric_features:
  data_new[feature] = pd.to_numeric(data_new[feature].replace('?', pd.NA), errors='coerce')
  data_new[feature].fillna(data_new[feature].mean(), inplace=True)
categorial = ['num-of-doors']
for feature in categorial:
  data_new[feature] = data_new[feature].replace('?', pd.NA)
  data_new[feature].fillna(data_new[feature].mode()[0], inplace=True)
data_new
```

Пропуски в столбце `normalized-losses` предскажите при помощи линейной регрессии по признакам
